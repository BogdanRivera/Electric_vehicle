In this notebook we show some data from a dataset of Electric Vehicles Population.


```python
#Import libraries
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
#Open CSV to work
archive = pd.read_csv("/Electric_Vehicle_Population_Data.csv")
#Convert in dataframe
df = pd.DataFrame(archive)
df.head()
```

We can group the model in model year and compare the electric range


```python
x_electric_range = df.groupby("Model Year")
x_electric_range.head()
```

To view some data of dataset we plot the model year and electric range and then, we group by 'Make' that is referenced to fabricant.

```python
plt.style.use('grayscale')
```

```python
sns.catplot(data=df,x='Model Year',y = "Electric Range",hue='Make',height = 10)
plt.show()
```

In the graphic we can see that company with more electric range is Tesla. The most efficient models are created in 2020. \
In addition we can how much electric vehicles are using existing in each country.

```python
vehicles_per_country = df.groupby("Country")['Make'].count()
vehicles_per_country
```

In the same way you can graph the relationship of manufacturers and electrical autonomy grouped by country



```python
sns.displot(data=df,x='Make',y ='Electric Range',height = 10,hue = 'Country')
plt.xticks(rotation=45)
plt.show()
```
